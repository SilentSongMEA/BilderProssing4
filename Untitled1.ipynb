```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread(r"C:\Users\Student\Desktop\pic\opencv_inv.png",0)
kernel=np.ones((5,5),np.uint8)
```

```python
erosion=cv2.erode(image,kernel,iterations=1)
dilation=cv2.dilate(image,kernel,iterations=1)
opening=cv2.morphologyEx(image,cv2.MORPH_OPEN,kernel)
closing=cv2.morphologyEx(image,cv2.MORPH_OPEN,kernel)
plt.figure(figsize=[10,8])
plt.subplot(231);plt.imshow(image,cmap="gray");plt.title("orginal")
plt.subplot(232);plt.imshow(erosion,cmap="gray");plt.title("erosion")
plt.subplot(233);plt.imshow(dilation,cmap="gray");plt.title("dilation")
plt.subplot(234);plt.imshow(opening,cmap="gray");plt.title("opening")
plt.subplot(236);plt.imshow(closing,cmap="gray");plt.title("closing")
plt.show()
```
