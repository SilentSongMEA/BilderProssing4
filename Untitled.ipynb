```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import ImageGrab
```

```python
while True:
    frame=np.array(ImageGrab.grab())
    cv2.imshow("screen",cv2.cvtColor(frame,cv2.COLOR_BGR2RGB))
    if cv2.waitKey(1) & 0XFF==27:
        break
cv2.destroyAllWindows()
```

```python
while True:
    frame=np.array(ImageGrab.grab(bbox=(100,100,1000,1000)))
    cv2.imshow("screen area",cv2.cvtColor(frame,cv2.COLOR_BGR2RGB))
    if cv2.waitKey(1) & 0XFF ==27:
        break
cv2.destroyAllWindows()
```

```python
image1=cv2.imread(r"C:\\Users\Student\Desktop\pic\noisy.jpg",0)
ret1,thresh1=cv2.threshold(image1,127,255,cv2.THRESH_BINARY)
ret2,thresh2=cv2.threshold(image1,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
blur=cv2.GaussianBlur(image1,(5,5),0)
ret3,thresh3=cv2.threshold(blur,127,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
images=[image1,0,thresh1,
       image1,0,thresh2,
       blur,0,thresh3]
title=["orginal noisy image","histogram","global thresh(127)",
       "orginal noisy image","histogram","otsu thresh",
       "gaussianBlur filltered image","histogram","otsu thresh"]
plt.figure(figsize=(15,7))
for i in range(3):
    plt.subplot(3,3,i*3+1),plt.imshow(images[i*3],"gray"),
    plt.title(title[i*3]),plt.xticks([]),plt.yticks([])
    
    plt.subplot(3,3,i*3+2),plt.hist(images[i*3].ravel(),256)
    plt.title(title[i*3+1]),plt.xticks([]),plt.yticks([])
    
    plt.subplot(3,3,i*3+3),plt.imshow(images[i*3+2],"gray")
    plt.title(title[i*3+2]),plt.xticks([]),plt.yticks([])

```

```python
orginal_image=cv2.imread(r"C:\\Users\Student\Desktop\pic\sudoku.jpg",0)
noised_remove=cv2.medianBlur(orginal_image,5)
_,output1=cv2.threshold(orginal_image,127,255,cv2.THRESH_BINARY)
output2=cv2.adaptiveThreshold(orginal_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
output3=cv2.adaptiveThreshold(orginal_image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

_,output4=cv2.threshold(noised_remove,127,255,cv2.THRESH_BINARY)
output5=cv2.adaptiveThreshold(noised_remove,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
output6=cv2.adaptiveThreshold(noised_remove,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

plt.figure(figsize=(15,15))
plt.subplot(331);plt.imshow(orginal_image,cmap="gray");plt.title("Orginal")
plt.subplot(331);plt.imshow(noised_remove,cmap="gray");plt.title("noised_remove")
plt.subplot(333);plt.imshow(output1,cmap="gray");plt.title("output1")
plt.subplot(334);plt.imshow(output2,cmap="gray");plt.title("output2")
plt.subplot(335);plt.imshow(output3,cmap="gray");plt.title("output3")
plt.subplot(336);plt.imshow(output4,cmap="gray");plt.title("output4")
plt.subplot(337);plt.imshow(output5,cmap="gray");plt.title("output5")
plt.subplot(338);plt.imshow(output6,cmap="gray");plt.title("output6")
```
